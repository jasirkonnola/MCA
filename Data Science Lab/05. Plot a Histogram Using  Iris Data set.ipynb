```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

df = sns.load_dataset('iris')
fig, ax = plt.subplots(1, 1)
ax.hist(df['petal_length'], bins=49, edgecolor='black')
ax.set_title("Histogram")
ax.set_xlabel('Petal Length')
ax.set_ylabel('Frequency')
plt.show()
```

```python
from matplotlib import pyplot as plt
import pandas as pd

# Load the dataset (use raw string to avoid path errors)
df = pd.read_csv(r'C:\Users\Dell\Downloads\iris.csv')

# Inspect column names (optional but useful for debugging)
print("Available columns:", df.columns)

# Create the plot
fig, ax = plt.subplots(1, 1)

# Plot histogram of petal length
ax.hist(df['PetalLengthCm'], bins=49, edgecolor='black')

# Customize the plot
ax.set_title("Histogram of Petal Length")
ax.set_xlabel('Petal Length (cm)')
ax.set_ylabel('Frequency')
ax.set_xticks([1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5, 7.0])
ax.set_yticks(range(0, 20))

# Show the plot
plt.show()
```
